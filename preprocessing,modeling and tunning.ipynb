```python
fit.columns=['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']
fit
```

```python
fit[['City','Date']]=df[['City','Date']]
fit
```

```python
def custom_imputer(fit):
    if fit['AQI'] < 51.0:
        return 'Good'
    elif 50.0<fit['AQI']<101.0:
        return 'Satisfactory'
    elif 100.0<fit['AQI']<201.0:
        return 'Moderate'
    elif 200.0<fit['AQI']<301.0:
        return 'Poor'
    elif 300.0<fit['AQI']<401.0:
        return 'Very Poor'
    else:
        return 'Severe'

fit['AQI_Bucket'] = fit.apply(custom_imputer,axis=1)
```

# VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X=fit[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']]

calc_vif(X)
```

No VIF values is greater than 5 so we can assume moderate multicollinearity

```python
from collections import Counter
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

# Oversampling with SMOTE

```python
# summarize class distribution
print("Before oversampling: ",Counter(y_train))

# define oversampling strategy
SMOTE = SMOTE()

# fit and apply the transform
X_train_SMOTE, y_train_SMOTE = SMOTE.fit_resample(X_train, y_train)

# summarize class distribution
print("After oversampling: ",Counter(y_train_SMOTE))
```

```python
print("Accuracy is: ", accuracy_score(y_test,Y_predran))
print("Precision is: ", precision_score(y_test,Y_predran,  average = 'micro'))
print("Recall is: ", recall_score(y_test,Y_predran,  average = 'micro'))
print("F1 score is: ", f1_score(y_test,Y_predran, average = 'micro'))
```
